```python
import pandas as pd # imported numpy and pandas
import numpy as np
```

```python
from pandas.io.html import read_html   # import library to read html pages
page="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
table= read_html(page, attrs={"class":"wikitable"})  # reading table from html page
```

```python
print("extacted {num} table".format(num=len(table)))  # printing number of table got extracted
```

```python
table[0].head()
```

```python
df=pd.DataFrame(table[0]) # Assign table to dataframe df
df.rename(columns={'Postal code':'PostalCode'},inplace=True)
df.head()
```

```python
df1=df[df.Borough !='Not assigned'] # deleted all not "assigned code" of borough column
df2=df1
```

```python
df2['Neighborhood']=df1['Neighborhood'].str.split('/')  # split the column value by '/' and assingn to df2 dataframe
```

```python
df2.head()
```

```python
df2.reset_index(drop=True, inplace=True)
```
